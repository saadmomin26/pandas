```python

import pandas as pd

data={
    "Name":['Arun','varu ','karun','Narun',"Marun"],
    "Age":[28,34,22,34,28],
    "Salary":[50000,60000,45000,52000,48000]
}
df=pd.DataFrame(data)
print(df)
```

```python
#grouping
grouped=df.groupby("Age")["Salary"].sum()
print(grouped)
```

```python
# grouping multi columen
grouped = df.groupby(["Age", "Name"])["Salary"].sum()
print(grouped)
```

```python
sum() → Total

mean() → Average

min() → Minimum

max() → Maximum

count() → Count of values

std() → Standard deviation

var() → Variance

median() → Middle value
```
