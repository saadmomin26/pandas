```python
import pandas as pd
data={
    "Name":['sham',"ram",'ghanshyam','dhanshyam','alice','bob','raj','simarn'],
    "Age":[10,None,28,23,33,42,24,26],
    "Salary":[50000,None,450000,52000,56000,46000,70000,85000],
    "Performace Score":[85,None,66,90,92,55,70,89]
 }
df=pd.DataFrame(data)
print("sample Dataframe")
print(df)

```

```python
#interpolate method
#It fills missing (NaN) values in the DataFrame by estimating values in a straight line between existing data points, column-wise.
df.interpolate(method="linear",axis=0,inplace=True)
print(df)
```
