```python
import pandas as pd
data={
    "Time":[1,2,3,4,5],
    "Value":[10,None,30,40,50]
}  
df=pd.DataFrame(data)
print("sample Dataframe")
print(df)

#The linear method fills missing values by assuming the data changes at a constant rate between two known values.
df["Value"]=df["Value"].interpolate(method="linear")
print("after interpolate")
print(df)
```
