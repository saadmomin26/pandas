```python
import pandas as pd

data={
    "Name":['Arun','varu ','karun'],
    "Age":[28,34,22],
    "Salary":[10000,20000,30000]
}
df=pd.DataFrame(data)
print(df)

#sorting one column
df.sort_values(by="Age", ascending=False,inplace=True)
print(df)
```

```python
#sorting multi column
df.sort_values( by=["Age","Salary"],ascending=[False,True],inplace=True)
print(df)

```
