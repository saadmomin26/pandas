```python
import pandas as pd
data={
    "Name":['sham','ram','ghanshyam','dhanshyam','alice','bob','raj','simarn'],
    "Age":[10,20,30,23,33,42,24,26],
    "Salary":[50000,60000,450000,52000,56000,46000,70000,85000],
    "performace Score":[85,55,66,90,92,55,70,89]
 }
df=pd.DataFrame(data)
print("sample Dataframe")
print(df)

#single row acces
rows=df[df['Salary']>=50000 ]
print("Employee with Salary> 50000")
print(rows)

#multi row acces
multirows=df[(df['Age']>30)&(df['Salary']>=50000) ]
print("Employee list age> 30 + Salary> 50000")
print(multirows)
```

```python
multirows2 = df[(df['Age'] > 30) | (df['performace Score'] >90)]
print("Employee list age > 30 or performace score>90")
print(multirows2)
```
