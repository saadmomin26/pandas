```python
import pandas as pd
data={
    "Name":['sham','ram','ghanshyam'],
    "age":[10,20,30,],
    "city":['mumbai','dheli','nagpur']
 }
df=pd.DataFrame(data)
print(df)
```

```python
df.to_csv("output.csv",index=False)  #it is used to save file in csv
```

```python
df.to_excel("output.xlsx",index=False)  #it is used to save file in excel
```

```python
df.to_json("output.json",index=False)  #it is used to save file in json
```
