```python
import pandas as pd
data={
    "Name":['sham',None,'ghanshyam','dhanshyam','alice','bob','raj','simarn'],
    "Age":[10,None,28,23,33,42,24,26],
    "Salary":[50000,None,450000,52000,56000,46000,70000,85000],
    "Performace Score":[85,None,66,90,92,55,70,89]
 }
df=pd.DataFrame(data)
print("sample Dataframe")
print(df)

# it will show the missing data in colume
print(df.isnull())
#count of missing data
print(df.isnull().sum())


#filling missing data
df.fillna(0,inplace=True)
print(df)
```

```python
#droping missing data
df.dropna(inplace=True)
print(df)
```

```python
import pandas as pd
data={
    "Name":['sham',None,'ghanshyam','dhanshyam','alice','bob','raj','simarn'],
    "Age":[10,None,28,23,33,42,24,26],
    "Salary":[50000,None,450000,52000,56000,46000,70000,85000],
    "Performace Score":[85,None,66,90,92,55,70,89]
 }
df=pd.DataFrame(data)
print("sample Dataframe")
print(df)


#filling missing data of age by mean
df["Age"].fillna(df["Age"].mean(),inplace=True)
df["Salary"].fillna(df["Salary"].mean(),inplace=True)
print(df)

```
